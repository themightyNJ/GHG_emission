```python
import pandas as pd
```

```python
data = pd.read_csv(r"D:\@themightyNJ\DTU_AI\xclara.csv")
data.head()
```

```python
c1 = data["V1"].values
c2 = data["V2"].values
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.scatter(c1,c2)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
sse = {}
for k in range(1,11):
    kmeans = KMeans(n_clusters = k)
    kmeans_label = kmeans.fit(data)
    sse[k] = kmeans_label.inertia_
print(sse)
```

```python
plt.xlabel("<-Optimum K value->")
plt.ylabel("<-Difference->")
plt.title("'Elbow Method'")
plt.scatter(list(sse.keys()),list(sse.values()),c="red")
plt.plot(list(sse.keys()),list(sse.values()),c="black")
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3)
kmeans_label = kmeans.fit_predict(data)
```

```python
plt.scatter(c1,c2,c = kmeans.labels_)
plt.show()
```

```python
centroids = kmeans.cluster_centers_
print(centroids)
```
