```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Building a vector of 10000 normal deviates with variance 0.5^2 and mean 2
mu,sigma = 2,0.5
v = np.random.normal(mu,sigma,10000)
```

```python
#Plot a normalized histogram with 50 bins
plt.hist(v, bins = 50, density = 1,color = "#0ffff0")
plt.show()
```
