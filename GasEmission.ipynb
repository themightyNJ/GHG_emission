```python
totalEmission = data1["2017"].sum()
totalEmission2007 = data1["2007"].sum()
```

```python
pastList1 = [data1["1927"].sum(),data1["1937"].sum(),data1["1947"].sum(),data1["1957"].sum(),data1["1967"].sum(),data1["1977"].sum(),data1["1987"].sum(),data1["1997"].sum(),data1["2007"].sum(),data1["2017"].sum()]
plt.plot(pastList1)
plt.title("world in past 100 years")
plt.ylabel("Emission")
plt.xlabel("Time")
plt.show()
print("Total emission in 1927: ",data1["1927"].sum()/1000000,"Million Metric Tons")
print("Total emission in 1947: ",data1["1947"].sum()/1000000,"Million Metric Tons")
print("Total emission in 1967: ",data1["1967"].sum()/1000000,"Million Metric Tons")
print("Total emission in 1987: ",data1["1987"].sum()/1000000,"Million Metric Tons")
print("Total emission in 2007: ",data1["2007"].sum()/1000000,"Million Metric Tons")
print("Total emission in 2017: ",data1["2017"].sum()/1000000,"Million Metric Tons")
print("Increase in emission in past 10 years: ",round(totalEmission*100/totalEmission2007)-100,"% approximately")
print("Increase in emission in past 25 years: ",round(totalEmission*100/data1["1993"].sum())-100,"% approximately")
print("Increase in emission in past 50 years: ",round(totalEmission*100/data1["1967"].sum())-100,"% approximately")
print("Increase in emission in past 100 years: ",round(totalEmission*100/data1["1917"].sum())-100,"% approximately")
```

```python
champ = []
for i in range(1917,2018):
    m = data[str(i)].max()
    m1 = list(data[str(i)]).index(m)
    champ.append(Countries[m1])
    print("Maximum Emission in {} is Caused By: ".format(i),Countries[m1])
sns.countplot(champ)
plt.show()
```
