```python
df2 = pd.DataFrame({"Name":["Anshu","Saloni","Virat","Ashish"],
                   "Roll No":[5,6,7,8],
                   "Class":["DS","ML","ML","DS"],
                   "Marks":[40,40,40,40]},
                  index = [5,6,7,8])
```

```python
pd.concat([df1,df2])
```
